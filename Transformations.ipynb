```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Translation

```python

path = "misc/"
imgpath1 =  path + "4.2.01.tiff"
img1 = cv2.imread(imgpath1, 1)
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)


rows, columns, channels = img2.shape

T = np.float32([[1, 0, -150], [0, 1, -50]])

print(T)


output = cv2.warpAffine(img2, T, (columns, rows))


plt.imshow(output)
plt.title('Shifted Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

# Rotation

```python

rows, columns, channels = img1.shape
img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

R = cv2.getRotationMatrix2D((columns/2, rows/2), 370, 1)

print(R)


output = cv2.warpAffine(img2, R, (columns, rows))


plt.imshow(output)
plt.title('Rotated Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

# Affine Transformation

```python

point1 = np.float32([[100, 100], [300, 100], [100, 300]])
point2 = np.float32([[200, 150], [600, 150], [400, 300]])

A = cv2.getAffineTransform(point1, point2)
print(A)


output = cv2.warpAffine(img2, A, (columns, rows))


plt.imshow(output)
plt.title('Transformed Image')
plt.xticks([])
plt.yticks([])
plt.show()
```
