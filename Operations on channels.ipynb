```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

# Splitting of color channels

```python


path = "misc/4.2.03.tiff"

img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r, g, b = cv2.split(img)
titles = ['Original', 'Red', 'Blue', 'Green']
images = [cv2.merge((r, g, b)), r, g, b]

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(images[0], cmap='gray')
plt.title(titles[0], fontsize=20)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.imshow(images[1], cmap='Reds')
plt.title(titles[1], fontsize=20)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 3)
plt.imshow(images[2], cmap='Greens')
plt.title(titles[2], fontsize=20)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 4)
plt.imshow(images[3], cmap='Blues')
plt.title(titles[3], fontsize=20)
plt.xticks([])
plt.yticks([])

plt.show()
```

# Max RGB filter
-  Sets all the pixel values of each channel(which are less than the maximum pixel value across all channels for that particular pixel) to 0 

```python

path = "misc/Fig0635(top_ left_flower).tif"

img = cv2.imread(path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

R, G, B = cv2.split(img)
M = np.maximum(np.maximum(R, G), B)

R[R < M] = 0
G[G < M] = 0
B[B < M] = 0

plt.figure(figsize=(10, 10))
output = cv2.merge((R, G, B))
plt.imshow(output)
plt.title("Max RGB filter", fontsize=20)
plt.xticks([])
plt.yticks([])
plt.show()
```
