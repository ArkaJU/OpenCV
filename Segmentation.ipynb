```python
import cv2
import matplotlib.pyplot as plt
```

#                       Thresholding and Basic Segmentation    

```python

path = 'misc/'
imgpath1 = path + '7.1.02.tiff'
img = cv2.imread(imgpath1, 1)

th = 127     #threshold
maxval = 255
ret, o1 = cv2.threshold(img, th, maxval, cv2.THRESH_BINARY)
ret, o2 = cv2.threshold(img, th, maxval, cv2.THRESH_BINARY_INV)
ret, o3 = cv2.threshold(img, th, maxval, cv2.THRESH_TOZERO)
ret, o4 = cv2.threshold(img, th, maxval, cv2.THRESH_TOZERO_INV)
ret, o5 = cv2.threshold(img, th, maxval, cv2.THRESH_TRUNC)

output = [img, o1, o2, o3, o4, o5]
titles = ['Original', 'Binary', 'Binary Inv', 'Zero', 'Zero Inv', 'Trunc' ]

plt.figure(figsize=(15, 15))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
    
```

# Otsu's binarization thresholding 
 -   Selects theshold automatically
 -   Works well with bimodal images
 -   Bimodal: foregroud + background information

```python
path = 'misc/'
imgpath1 = path + '7.1.08.tiff'
img = cv2.imread(imgpath1, 0)
th = 0
maxval = 255
ret, o1 = cv2.threshold(img, th, maxval, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
ret, o2 = cv2.threshold(img, th, maxval, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
ret, o3 = cv2.threshold(img, th, maxval, cv2.THRESH_TOZERO + cv2.THRESH_OTSU)
ret, o4 = cv2.threshold(img, th, maxval, cv2.THRESH_TOZERO_INV + cv2.THRESH_OTSU)
ret, o5 = cv2.threshold(img, th, maxval, cv2.THRESH_TRUNC + cv2.THRESH_OTSU)

output = [img, o1, o2, o3, o4, o5]
titles = ['Original', 'Binary', 'Binary Inv', 'Zero', 'Zero Inv', 'Trunc' ]

plt.figure(figsize=(15, 15))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(output[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

# Adaptive Thresholding
  -  Works well with with images having different lighting conditions

```python
path = 'misc/'
imgpath1 = path + '5.1.12.tiff'
img = cv2.imread(imgpath1, 0)
blocksize = 513
c = 2
maxval = 255
th1 = cv2.adaptiveThreshold(img, maxval, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, blocksize, c)
th2 = cv2.adaptiveThreshold(img, maxval, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blocksize, c)

output = [img, th1, th2]
titles = ['Original', 'Mean Adaptive', 'Gaussian Adaptive']

plt.figure(figsize=(15, 15))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(output[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
